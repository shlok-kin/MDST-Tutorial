```python
# @@@ Queston 6: fizzbuzz
# you will need to use both iteration and control flow 
# go through all numbers from 1 to 30 in order
# if the number is a multiple of 3, print fizz
# if the number is a multiple of 5, print buzz
# if the number is a multiple of 3 and 5, print fizzbuzz and NOTHING ELSE
# if the number is neither a multiple of 3 nor a multiple of 5, print the number
for i in range(1, 31):
    if i % 3 == 0 and i % 5 == 0:
        print("fizzbuzz")
    elif i % 3 == 0:
        print("fizz")
    elif i % 5 == 0:
        print("buzz")
    else:
        print(i)
```

```python
# @@@ Question 8: load in the "starbucks.csv" dataset
# refer to how we read the cereal.csv dataset in the tutorial
import pandas as pd
starbucks_df = pd.read_csv('starbucks.csv')
```

```python
# @@@ Question 9: select all rows with more than and including 400 calories
high_cal_drinks = starbucks_df[starbucks_df['calories'] >= 400]
print(high_cal_drinks)
```

```python
# @@@ Question 10: select all rows whose vitamin c content is higher than the iron content
vitamin_c_higher_than_iron = starbucks_df[starbucks_df['vitamin c'] > starbucks_df['iron']]
print(vitamin_c_higher_than_iron)
```

```python
# @@@ Question 11: create a new column containing the caffeine per calories of each drink
starbucks_df['Caffeine per Calorie'] = starbucks_df['caffeine'] / starbucks_df['calories']
print(starbucks_df[['beverage', 'Caffeine per Calorie']])
```

```python
# @@@ Question 12: what is the average calorie across all items?
average_calories = starbucks_df['calories'].mean()
print(f"Average calories across all items: {average_calories}")
```

```python
# @@@ Question 13: how many different categories of beverages are there?
num_categories = starbucks_df['beverage_category'].nunique()
print(num_categories)
```

```python
# @@@ Question 14: what is the average # calories for each beverage category?
average_calories_per_category = starbucks_df.groupby('beverage_category')['calories'].mean()
print(average_calories_per_category)
```
